Tarea 3.8 Un sensor recibe en promedio $50$ partículas por minuto. Ese
sensor está diseñado tal que puede detectar si fue impactado por al
menos una partícula, pero no tiene ninguna información sobre cuántas
partículas lo impactaron.

El sensor se encendió durante un segundo e indicó impacto. Grafique la
distribución de probabilidad de que ese pixel haya sido impactado por
$n$ partículas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

```python
#Vamos a usar la distribución de Poisson
sigma = 50/60 #cincuenta particulas en sesenta segundos
def poisson(n, sigma):
  return sigma**n*np.exp(-sigma)/factorial(n)

#Ahora usamos la función de probabilidad
def prob(n, sigma):
  if n==0:
    return 0
  else:
    return poisson(n, sigma)/(1-poisson(0, sigma))
#Ahora le asignamos valores a n
n_valores = np.arange(1, 11) #Va del 1 al 10
probabilidad = [prob(n, sigma) for n in n_valores]

#Ahora se grafica la probabilidad de que n particulas impacten el sensor
plt.figure(figsize=(8,6))
plt.bar(n_valores, probabilidad, edgecolor='black')
plt.xlabel('Numero de particulas (n)')
plt.ylabel('Probabilidad de impacto')
plt.title('Probabilidad de que n particulas impacten el sensor')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show
```
