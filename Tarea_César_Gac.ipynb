Tarea 1.9
Suponga que usted es un ayudante de un laboratorio de física. En un experimento se pide medir la posición de una masa que cae en un plano inclinado a intervalos regulares en el tiempo. El equipo disponible puede medir las posiciones con una precisión de 4mm. Uno de los estudiantes usa los siguientes datos (en cm):

datos = [0.9,  2. ,  3.6,  5.6,  8. , 10.9, 14.2, 18.]

Grafique estos datos con sus barras de error. Explique por qué el estudiante merece una nota de 1.0. Compare con los siguientes datos obtenidos por un estudiante que merece 7.0.

datos_7 = [0.3,  1.7,  4.1,  5.7,  8.5, 10.7, 13.6, 17.2]

```python
import numpy as np
import matplotlib.pyplot as plt

#Datos de posiciones en cm
datos = [0.9,  2. ,  3.6,  5.6,  8. , 10.9, 14.2, 18.]
datos_7 = [0.3,  1.7,  4.1,  5.7,  8.5, 10.7, 13.6, 17.2]

#Tiempo en s, por intervalos regulares
tiempo = np.linspace(0,7,8)

#Error en cm
error = 0.4

#Se harán dos graficos
fig = plt.figure()

axs1 = fig.add_subplot(1,2,1)
axs2 = fig.add_subplot(1,2,2)

#Se grafican los datos del primer estudiante
axs1.errorbar(tiempo, datos, yerr=error, color='red', label='Datos con error 4mm')
#Se garfican los datos del segundo estudiante
axs2.errorbar(tiempo, datos_7, yerr=error, color='green', label='Datos con error 4mm')

#Se añaden los texto al gráfico
fig.suptitle('Graficos de datos de dos alumnos')
axs1.set_title('Datos Primer alumno')
axs2.set_title('Datos Segundo Alumno')
axs1.set_xlabel('Tiempo')
axs1.set_ylabel('Posición')
axs2.set_xlabel('Tiempo')
axs2.set_ylabel('Posición')
axs1.legend()
axs2.legend()

#Se muestra los graficos
plt.tight_layout()
plt.show
```

En ambos gráficos se puede apreciar que los datos tienden a una curva, la cual refleja la aceleración que experimenta la masa mientras más tiempo está cayendo. El gráfico del primer alumno posee una curva más "suave" y perfecta que el segundo, sin embargo, aunque parezca contraintuitivo el grafico del segundo alumno posee datos más consistentes, pues existen diversas variables que pueden inducir pequeños errores, como la presición, el tiempo de reacción, el mismo roce con el aire, etc. Por lo que el grafico del primer estudiante es sospechoso y es bastante probable que los datos hubieran sido modificados para que la curva se viera lo más cercano a "perfecta", por lo que tendria una nota 1 por la manipulación de los datos, mientras que los datos del segundo alumno son más consitentes y tendria una nota 7

# Nota: 7.0
